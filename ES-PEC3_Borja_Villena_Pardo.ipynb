Pasos de función **approximate_pi()**:

```python
# Hemos usado pycodestyle para algunas comprobaciones esporádicas.
# Lo desactivamos porque nos origina algunos errores al ejecutar el
# código
'''
%load_ext pycodestyle_magic
# %pycodestyle_on
'''
```

```python
# Importamos librerías necesarias

import matplotlib.colors as mcolors
import matplotlib.pyplot as plt


# Copiamos función approximate_pi()

def approximate_pi(N):
    value = 0
    for k in range(1, 2*N+1, 2):
        sign = -(k % 4 - 2)
        value += float(sign) / k
    return 4 * value


# Definimos los valores de N (número de sumandos) que queremos evaluar
N_value = [10, 100000, 250000, 500000, 750000, 900000, 1000000]


# Creamos función time_pi_grafica para realizar la gráfica de los
# tiempos de ejecución


def time_pi_grafica(N):

    """
    Plot de tiempo de ejecución de la función approximate_pi()
    en función del valor N dado.
    """

# Creamos lista vacía dónde iremos incluyendo los valores de los
# tiempos calculados

    time_ejecucion = []
    
# Usamos loop 'for' con método 'time' para calcular los tiempos
# de ejecución de la función 'approximate_pi()' para cada valor de N dado.

    for i in N:
        start = timeit.default_timer()
        approximate_pi(i)
        stop = timeit.default_timer()
        time = stop - start
        time_ejecucion.append(time)
        
# Usamos matplotlib para crear el gráfico
# Definimos parámetros y customización de la línea
    plt.figure(figsize=(10,6))
    plt.plot(N_value, time_ejecucion, 'k', marker='o', mfc='r')
    
# Definimos los valores de N en el eje X
    plt.xticks(N)
    
# Definimos título al gráfico y ejes.
    plt.title('Complejidad asintótica del algoritmo')
    plt.xlabel('Número de sumandos N')
    plt.ylabel('Tiempo de ejecución en segundos')
    
# Añadimos grid de fondo e imprimimos gráfico
    plt.grid(True)
    plt.show()


# Llamamos a la función para crear la gráfica. Cómo podemos
# comprobar, el tiempo aumenta respecto al número de sumandos
# de manera lineal.
time_pi_grafica(N_value)
```

```python
# Implementamos método Linear Search encontrado en //geeksforgeeks.org
# con algunos retoques para comprobar cifras de 7 números.

def linearsearch(arr, N, x): 
    for i in range(0, N): 
        if (arr[i] == x):
            return i
        elif len(str(arr[i])) < x:
            return -2
    return -1

# Aplicamos método assert para primer elemento de la lista
assert linearsearch(lista_6_int_sinDU, len(lista_6_int_sinDU), 141592) == 0

# Aplicamos método assert para demostrar ausencia de cifra de 
# 7 números
assert linearsearch(lista_6_int_sinDU, len(lista_6_int_sinDU), 7) == -2

# Implementamos método Sentinel Linear Search encontrado en //geeksforgeeks.org
# con algunos retoques para que nos retorne un valor u otro según corresponda.

def sentinelSearch(arr, n, key):
    last = arr[n - 1]
    arr[n - 1] = key
    i = 0
    while (arr[i] != key):
        i += 1
    arr[n - 1] = last
    if ((i < n - 1) or (arr[n - 1] == key)):
        return i
    
# Añadimos siguiente elif para el segundo assert
    elif len(str(arr[i])) < key:
        return -2
    else:
        return -1

# Aplicamos método assert para primer elemento de la lista

assert sentinelSearch(lista_6_int_sinDU, len(lista_6_int_sinDU), 141592) == 0

# Aplicamos método assert para demostrar ausencia de cifra de 
# 7 números

assert sentinelSearch(lista_6_int_sinDU, len(lista_6_int_sinDU), 7) == -2
```

```python
# Importamos librerias

import math

# Codificamos código de método MetabinarySearch optenido de //geeksforgeeks.org
# con algunas modificaciones en los types

def meta_binary_search(A, key_to_search):
    n = len(A)
    lg = int(math.log2(n-1)) + 1;
    pos = 0
    for i in range(lg - 1, -1, -1) :
        if (A[pos] == key_to_search):
            return pos
        new_pos = pos | (1 << i)
        if ((new_pos < n) and
            (A[new_pos] <= key_to_search)):
            pos = new_pos
    return (pos if(A[pos] == key_to_search) else -1)

# Codificamos código de método TernarySearch optenido de //geeksforgeeks.org
# con algunas modificaciones en los types

def ternarySearch(l, r, elemento, lista):
    if (r >= l):
        mid1 = int(l + (r - l) / 3)
        mid2 = int(r - (r - l) / 3)
        if (lista[mid1] == elemento): 
            return mid1
        if (lista[mid2] == elemento): 
            return mid2
        if (elemento < lista[mid1]): 
            return ternarySearch(l, mid1 - 1, elemento, lista)
        elif (elemento > lista[mid2]): 
            return ternarySearch(mid2 + 1, r, elemento, lista)
        else: 
            return ternarySearch(mid1 + 1, 
                                 mid2 - 1, elemento, lista)
    return -1

# Creamos función para comparar los tres métodos que nos solicitan

def binaryVSmetaVSternary(lista, longitud = None, primero = None, 
                        ultimo = None):
    """
    Plot de tiempo de ejecución de los algoritmos
    BinarySearch, MetabinarySearch y TernarySearch
    """
# Definimos valores de parámetros por defecto

    if longitud is None:
        longitud = len(lista)
        
    if primero is None:
        primero = lista[0]
        
    if ultimo is None:
        ultimo = lista[-1]

# Creamos dos listas para guardar los tiempos de ejecución de ambas

    time_binary = []
    time_meta = []
    time_ternary = []

# Usamos timeit.repeat() tal y como nos pide el enunciado. Nos hemos
# apoyado en //openai.com para asegurar buen planteamiento de función
# lambda.

    for i in lista:
        time_binary.append(timeit.repeat(lambda: binarySearch(lista, 0, len(lista) - 1, i), 
                                         repeat = 10, number = 1000))
        time_meta.append(timeit.repeat(lambda: meta_binary_search(lista, i), 
                                         repeat = 10, number = 1000))
        time_ternary.append(timeit.repeat(lambda: ternarySearch(0, len(lista) - 1, i, lista), 
                                         repeat = 10, number = 1000))

# Calculamos la media para suavizar los resultados
    avg_time_binary = np.mean(time_binary, axis=1)
    avg_time_meta = np.mean(time_meta, axis=1)
    avg_time_ternary = np.mean(time_ternary, axis=1)
    
# Usamos matplotlib para crear el gráfico
# Definimos parámetros y customización de la línea
    plt.figure(figsize=(10,6))

# Imprimimos datos suavizados
    plt.plot(avg_time_binary, 'b', label = 'BinarySearch')
    plt.plot(avg_time_meta, 'k', label = 'MetabinarySearch')
    plt.plot(avg_time_ternary, 'r', label = 'TernarySearch')

# Definimos título al gráfico y ejes.
    plt.ylim(0, 0.02) 
    # Mientras mas aumentemos el valor del segundo parametro de 
    # plt.ylim() mejor se vera el comportamiento lineal 
    plt.title('Complejidad asintótica del algoritmo', y = 1.05, fontsize = 16)
    plt.xlabel('Número de elementos analizados en array', fontsize = 14)
    plt.ylabel('Tiempo de ejecución en $seg·10^-6$', fontsize = 14)

# Añadimos grid de fondo, leyenda e imprimimos gráfico
    plt.grid(True, color = 'green')
    plt.legend(loc = 'upper left', framealpha = 1, facecolor='lightgreen')
    plt.show()
    
# Imprimimos datos 
grafico_VS3 = binaryVSmetaVSternary(lq_10)
```

```python
# Importamos librerias

import decimal

# Codificamos función Chudnovski extraída de la url
# indicada: //levelup.gitconnected.com

def chudnovski(n):
    """
    Esta función calcula el valor de pi con 'n' numeros de decimales.
    """
    decimal.getcontext().prec = n + 3
    decimal.getcontext().Emax = 999999999
    
    C = 426880 * decimal.Decimal(10005).sqrt()
    K = decimal.Decimal(6)
    M = decimal.Decimal(1)
    X = decimal.Decimal(1)
    L = decimal.Decimal(13591409)
    S = L

    for i in range(1, n+3):
        M = decimal.Decimal(M* ((1728*i*i*i)-(2592*i*i)+(1104*i)-120)/(i*i*i))
        L = decimal.Decimal(545140134+L)
        X = decimal.Decimal(-262537412640768000*X)
        S += decimal.Decimal((M*L) / X) 
          
    return str(C/S)[:-2]

pi_100 = chudnovski(100)
print(f'El número \u03C0 con sus 100 primeros decimales es: \n{pi_100}')
```

En un proceso calcularemos los 7000 decimales del número pi. Y también crearemos un proceso para cada descarga de fichero. Igual que antes, queremos comparar la versión con un solo proceso (ya calculado) con la versión multiproceso. Calcula el tiempo total.

```python
# Importamos librerias
from multiprocessing import Process
import os

# Cambiamos de directorio para descargar/crear nuevos archivos
'''
Activamos esta línea de código en caso de haber sido ejecutada 
ya anteriormente por primera vez:

os.chdir('..') 
'''

os.chdir('../data/test/')

# Creamos funcion para descargar los archivos uno a uno

# Creamos funcion individual para descargar los archivos

def descargar_arch_individual(ruta):
    '''
    Función que descarga los archivos de las url dadas
    '''
    try:
        r = requests.get(ruta)

# Abrimos un archivo en nuestro sistema y escribimos 
# el contenido de los archivos descargados.

        if "pi.pdf" in ruta:
            with open("pi.pdf", "wb") as code:
                code.write(r.content)
                
        elif "pi-formulas.pdf" in ruta:  
            with open("pi-formulas.pdf", "wb") as code:
                code.write(r.content)
            
        elif "dhb-conant.pdf" in ruta:      
            with open("dhb-conant.pdf", "wb") as code:
                code.write(r.content)
            
        elif "1912.05319.pdf" in ruta:    
            with open("1912.05319.pdf", "wb") as code:
                code.write(r.content)
            
        elif "Monthly_Reference_12.pdf" in ruta:    
            with open("Monthly_Reference_12.pdf", "wb") as code:
                code.write(r.content)
            
    except Exception as e:
        print(f"Error en descargar_arch: {e}")

# Creamos variable time_ejecucion_nonlocal
time_ejecucion_nonlocal = []

# Creamos función para realizar los calculos en distintos procesos
def multiprocesos():
    '''
    Función que lanza multiprocesos
    '''
# Iniciamos timeit
    start = timeit.default_timer()

# Creamos los procesos
    procesos = []
    
# Proceso 1 para descargar archivos
    proceso_descargar_arch_1 = Process(target = descargar_arch_individual, 
                                       args = (url_1,))
    procesos.append(proceso_descargar_arch_1)
    proceso_descargar_arch_1.start()
    
# Proceso 2 para descargar archivos
    proceso_descargar_arch_2 = Process(target = descargar_arch_individual, 
                                       args = (url_2,))
    procesos.append(proceso_descargar_arch_2)
    proceso_descargar_arch_2.start()

# Proceso 3 para descargar archivos
    proceso_descargar_arch_3 = Process(target = descargar_arch_individual, 
                                       args = (url_3,))
    procesos.append(proceso_descargar_arch_3)
    proceso_descargar_arch_3.start()
    
# Proceso 4 para descargar archivos
    proceso_descargar_arch_4 = Process(target = descargar_arch_individual, 
                                       args = (url_4,))
    procesos.append(proceso_descargar_arch_4)
    proceso_descargar_arch_4.start()
    
# Proceso 5 para descargar archivos
    proceso_descargar_arch_5 = Process(target = descargar_arch_individual, 
                                       args = (url_5,))
    procesos.append(proceso_descargar_arch_5)
    proceso_descargar_arch_5.start()
    
# Proceso para ejecutar chudnovski
    proceso_chudnovski = Process(target = hilo_chudnovski)
    procesos.append(proceso_chudnovski)
    proceso_chudnovski.start()

# Esperamos a que ambos procesos terminen
    for proceso in procesos:
        proceso.join()

    stop = timeit.default_timer()
    time = stop - start
    time_ejecucion_nonlocal.append(time)

    print(f'El tiempo de ejecución total es: {time_ejecucion_nonlocal} segundos')

multiprocesos()
```
